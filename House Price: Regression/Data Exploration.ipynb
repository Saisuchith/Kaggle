# Building A Neural Network

```python
from keras.models import Sequential
from keras.layers import Dense,Dropout
```

```python
from sklearn.preprocessing import StandardScaler
sc_X=StandardScaler()
X_train_std=sc_X.fit_transform(X_train)
X_val_std=sc_X.transform(X_val)
X_test_std=sc_X.transform(X_test)


y_train_r=y_train.reshape(y_train.size,1)
y_val_r=y_val.reshape(y_val.size,1)
sc_y=StandardScaler()
y_train_std=sc_y.fit_transform(y_train_r)
y_val_std=sc_y.transform(y_val_r)

y_train_std=y_train_std.reshape(y_train.shape[0],)
y_val_std=y_val_std.reshape(y_val.shape[0],)
print(X_train_std.shape,X_val_std.shape,y_train_std.shape,y_val_std.shape)
```

```python
# defining the model
model=Sequential()
model.add(Dense(256,input_shape=(245,),activation='linear'))
model.add(Dense(64,activation='linear'))
model.add(Dense(32,activation='linear'))
model.add(Dense(1,activation='linear'))
model.summary()
```

```python
# compiling the model
model.compile(optimizer='adam',loss='mse',metrics=['mae'])

model.summary()
```

```python
model.fit(x=X_train_std,y=y_train_std,batch_size=50,epochs=30,validation_data=(X_val_std,y_val_std))
```

```python
nn_pred=model.predict(X_test_std)
print(nn_pred.shape)
```

```python
nn_pred=sc_y.inverse_transform(nn_pred)
nn_pred=nn_pred.reshape(1459,)
print(nn_pred.shape)
```

```python
nn=pd.DataFrame()
nn['Id']=id_test
nn['SalePrice']=nn_pred
nn.head(10)
```
